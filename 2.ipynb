```python
%matplotlib ipympl
import numpy as np
import matplotlib.pyplot as plt
from math import *
```

```python
with plt.ioff():
    fig, ax = plt.subplots()
```

```python
with plt.ioff():
    fig2, ax2 = plt.subplots()
```

```python
f = np.vectorize(lambda x: x*sin(2*pi*x))
x = np.linspace(-1, 1, 50)
y = f(x)
```

```python
N = 6
x_ep = np.random.uniform(-1, 1, N)
ep0 = .3
ep = np.random.uniform(-ep0, ep0, N)
y_ep = f(x_ep) + ep
ax.clear()
ax.plot(x, y, "b", label="x*sin(2*pi*x)")
ax.scatter(x_ep, y_ep, c="r")
ax.legend()
display(fig.canvas)
```

```python
A = np.vander(x_ep, N)
w = np.linalg.solve(A, y_ep)
ft = np.vectorize(lambda x: np.sum(np.power(x, np.arange(N)[::-1]) * w))
Er = np.sum((ft(x) - y)**2) / 2
print("Error: ", Er)
ax2.clear()
ax2.plot(x, y, "b", label="x*sin(2*pi*x)")
ax2.plot(x, ft(x), "g", label="f'")
ax2.scatter(x_ep, y_ep, c="r")
ax2.legend()
ax2.axis([x[0] - 1, x[-1] + 1, min(y) * 2, max(y) * 2])
display(fig2.canvas)
```
